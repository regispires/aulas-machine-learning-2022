## Entendendo os percentis

```python
a = range(0, 11)
```

```python
df_values = pd.DataFrame(a)
```

```python
df_values.describe()
```

- Os percentis são medidas que dividem a amostra (por ordem crescente dos dados) em 100 partes, cada uma com uma percentagem de dados aproximadamente igual.
- O 1º percentil determina o 1% menor dos dados
- O 98º percentil determina os 98% menores dos dados
- O 25º percentil é o primeiro quartil; o 50º percentil é a mediana. 
- O 10º percentil é o primeiro decil 
- Os dados são ordenados do menor ao maior e sua projeção é analisada.
