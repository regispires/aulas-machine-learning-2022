### Implementações das métricas que permitem receber dados do NumPy ou do Pandas

```python
def mae(y_true, y_pred):
    return np.sum(np.absolute(y_true - y_pred)) / y_true.shape[0]
```

```python
def mse(y_true, y_pred):
    return np.sum((y_true - y_pred)**2) / y_true.shape[0]
```

```python
def rmse(y_true, y_pred):
    return np.sqrt(np.sum((y_true - y_pred)**2) / y_true.shape[0])
```

#### Usandos dados do NumPy

```python
mae(y, y_pred)
```

```python
mae(df.y, df.y_pred)
```

```python
mse(y, y_pred)
```

```python
rmse(y, y_pred)
```

#### Usando dados do Pandas

```python
mae(df['y'], df['y_pred'])
```

```python
rmse(y, 4)
```

### Função para divisão entre treino e teste. 
- Pode receber dados do NumPy ou Pandas

```python
def my_train_test_split(X, y, train_ratio, seed):
    qtd = X.shape[0]
    idx = np.random.RandomState(seed=seed).permutation(np.arange(qtd))
    qtd_train = int(qtd * train_ratio)
    idx_train = idx[0:qtd_train]
    idx_test  = idx[qtd_train:]
    if type(X) == np.ndarray:    # NumPy Array
        X_train   = X[idx_train]
        X_test    = X[idx_test]
        y_train   = y[idx_train]
        y_test    = y[idx_test]
    else:                        # Pandas DataFrame
        X_train   = X.iloc[idx_train]
        X_test    = X.iloc[idx_test]
        y_train   = y.iloc[idx_train]
        y_test    = y.iloc[idx_test]
        
    return X_train, X_test, y_train, y_test
```
